```python
import matplotlib.pyplot  as plt
import random as rnd
```

```python
count=[0]*10
```

```python
world = [0] * 100
```

```python
for _ in range(100):
    x = rnd.randint(0,len(world)-1)
    p = rnd.randint(0,9)
    count[p]+=1
    world[x] = p
plt.bar([i for i in range(len(count))],count)
plt.show()
```

```python
count=[[]]*10
for i in range(20):
    for j in range(100):
        x= rnd.randint(0,len(world)-1)
        y= rnd.randint(0,len(world)-1)
        world[x]= world[y]
    cnt=[0]*10
    for k in range(len(world)):
        x = world[k]
        cnt[x]+=1
    print(cnt)
    for k in range(len(count)):
        count[k].append(cnt[k])
        print(count[k])

for i in range(10):
    plt.plot([i for i in range(len(count[i]))],count[i])
plt.show()        
```
